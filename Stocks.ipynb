# Moving average / rolling avg

A moving average is a tool used in finance to smooth out price data. Imagine trying to eliminate the noise from a stock’s price movements to create a clearer picture of where it’s heading. 

Moving averages are like a financial magnifying glass—they help us see the bigger picture behind stock price movements!

Within a week, a stock can swing up or down significantly. These data points are accurate, but they might not represent the overall trend over time.

Moving averages help provide a better idea of a stock’s price trends by calculating averages over different time periods.

# How Do Moving Averages Work?
Let’s say we’re looking at a 5-day moving average. Here’s how it’s calculated:
Add up the closing prices of the stock for the past 5 days.Divide the sum by 5 (since we’re considering 5 days).
Repeat this process for different time periods (e.g., 50-day or 200-day moving averages). Each data point on the graph represents the average of the prior days’ prices.

# Why Are Moving Averages Important?
Predictive Power: Moving averages allow analysts to predict potential stock directions before they happen. They’re considered lagging indicators because the information trails the stock’s price.

Smoothing Out Noise: By constantly updating the average price, moving averages help mitigate the impact of random short-term fluctuations.

Market Volatility: During volatile markets, moving averages provide more perspective and help traders make informed decisions.


```python
# Creating column 'moving avg of 50 and 200 day'
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
# Creating column 'previous day closing price' to see daily change in closing price
for df in dfs:
    df['Previous day closing price'] = df.Close.shift(1)
```

```python
apple.head()
# First entry is null as its first day of trading so theres 
# no day before closing price
```

```python
for df in dfs:
    df['Change in price'] = df['Close'] - df['Previous day closing price']
```

```python
apple.head()
# if -ve it decreased, if +ve it increased
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

# Volume

```python
for df in dfs:
    df['Change in volume'] = df['Volume'] - df['Previous day volume']
```
